```python
import pandas as pd
from chronos import BaseChronosPipeline
```

```python
pipeline = BaseChronosPipeline.from_pretrained("amazon/chronos-2", device_map="cpu")
```

```python
context_df = pd.read_excel("../data/cpg_inventory_60months.xlsx")
```

```python
context_df = context_df[["Product_ID", "Period", "Units_Sold"]]
```

```python
pred_df = pipeline.predict_df(
    context_df,
    prediction_length=1,  # Number of steps to forecast
    quantile_levels=[0.1, 0.5, 0.9],  # Quantiles for probabilistic forecast
    id_column="Product_ID",  # Column identifying different time series
    timestamp_column="Period",  # Column with datetime information
    target="Units_Sold",  # Column(s) with time series values to predict
)
```
